```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df
```

```python
tf = pd.read_csv("test.csv")
tf
```

- delete passenger id, name


```python
target = df["Transported"]
df = df.drop(["PassengerId","Name","Transported"],axis=1)
df
```

```python
passenger_id = tf["PassengerId"]
tf = tf.drop(["PassengerId","Name"],axis=1)
tf
```

```python
# check for null values
df.isnull().any()
```

```python
tf.isnull().any()
```

```python
# fill null values with -1
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="most_frequent",missing_values=np.nan)
df[df.columns] = imputer.fit_transform(df[df.columns])
df
```

```python
df.isnull().any()
```

```python
tf[tf.columns] = imputer.transform(tf[tf.columns])
tf
```

```python
# convert boolean statement to int
df["CryoSleep"] = df["CryoSleep"].apply(int)
df["VIP"] = df["VIP"].apply(int)
target = target.apply(int)
df
```

```python
tf["CryoSleep"] = tf["CryoSleep"].apply(int)
tf["VIP"] = tf["VIP"].apply(int)
tf
```

```python
def fl(x):
  return str(x)[0]
# create new columns, deck and side
df["Deck"] = df["Cabin"].apply(fl)
df
```

```python
tf["Deck"] = tf["Cabin"].apply(fl)
tf
```

```python
def side(x):
  return str(x)[-1]

df["Side"] = df["Cabin"].apply(side)
df
```

```python
tf["Side"] = tf["Cabin"].apply(side)
tf
```

```python
df = df.drop(["Cabin"],axis=1)
tf = tf.drop(["Cabin"],axis=1)
df,tf
```

```python
# apply label encoding
from sklearn.preprocessing import LabelEncoder

column_to_be_encoded = ["HomePlanet","Destination","Deck","Side"]

encoder = LabelEncoder()

df["HomePlanet"] = encoder.fit_transform(df["HomePlanet"])
tf["HomePlanet"] = encoder.transform(tf["HomePlanet"])
df["Destination"] = encoder.fit_transform(df["Destination"])
tf["Destination"] = encoder.transform(tf["Destination"])
df["Deck"] = encoder.fit_transform(df["Deck"])
tf["Deck"] = encoder.transform(tf["Deck"])
df["Side"] = encoder.fit_transform(df["Side"])
tf["Side"] = encoder.transform(tf["Side"])


df, tf
```

```python
target
```

```python
# build ml model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train,X_test,y_train,y_test = train_test_split(df,target,test_size=0.2,random_state=42)

X_train,X_test,y_train,y_test
```

```python
model = RandomForestClassifier()

model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
preds
```

```python
from sklearn.metrics import accuracy_score

acc_score = accuracy_score(y_test,preds)
acc_score
```

```python
tf
```

```python
test_preds = model.predict(tf)
test_preds
```
