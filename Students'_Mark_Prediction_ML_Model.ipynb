```python
# importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv file
df = pd.read_csv("Student_Marks.csv")
# display top 5 student
df.head()
```

```python
# seperate the data into x (independent variable) and y (dependent variable)
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
# use 20% of the data as test data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# create regression model
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# create prediction using test set
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
# print prediction and actual value side by side
comparison = np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_pred),1)),1)
            
print(comparison)
```

```python
# add grade column
def determineGrade(row):
  if row.Marks <= 100 and row.Marks >= 85:
    return 'A'
  elif row.Marks <= 84 and row.Marks >= 74:
    return 'B' 
  elif row.Marks <= 73 and row.Marks >= 60:
    return 'C' 
  elif row.Marks <= 59 and row.Marks >= 45:
    return 'D' 
  else:
    return 'F'
df["Grade"] = df.apply(determineGrade, axis="columns")

df
```
